# SepalLength

```python
sepal_length = iris[1:,0]
sepal_length
```

```python
np.nanmean(sepal_length)       #mean for sepal_length
```

```python
m = np.nanmedian(sepal_length)      #median
m
```

```python
a = np.where(np.isnan(sepal_length),np.nanmedian(sepal_length),sepal_length)  # replace nan with nanmedian
a
```

```python
ou = a > 8          # outlier
```

```python
sepal_length = np.where(ou,5,a)
sepal_length
```

```python
sepal_length = np.array(sepal_length,dtype = "float16")
sepal_length
```
